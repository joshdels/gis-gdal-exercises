```python
import pandas as pd
import matplotlib.pyplot as plt

#Load and check data
data = pd.read_csv("data/ahp_union_report.csv")
print(data.head())
df = pd.DataFrame(data)
```

```python
#Pivot
pivot_df = df.pivot(index='ADM4_EN', columns='level_v2', values='area_m2')
pivot_df = pivot_df[['Low', 'Medium', 'High']]

#Plot
pivot_df.plot(kind='bar', figsize=(12,6))

plt.suptitle('Suitable Area Designation (AHP + CIVAT)', fontsize=16, fontweight='bold', ha='center')
plt.xlabel("Barangay")
plt.ylabel('Area (m)')
plt.legend(title='Level')
plt.tight_layout()
plt.show()
```

```python
def data_bargraph(data_source, type, level):
  data = pd.read_csv(data_source)
  print(data.head())
  df = pd.DataFrame(data)
  pivot_df = df.pivot(index='ADM4_EN', columns=level, values='area_m2')
  pivot_df = pivot_df[['Low', 'Medium', 'High']]

  #Plot
  pivot_df.plot(kind='bar', figsize=(12,6))

  plt.suptitle(f'Suitable Area Designation ({type}+ CIVAT)', fontsize=16, fontweight='bold', ha='center')
  plt.xlabel("Barangay")
  plt.ylabel('Area (m)')
  plt.legend(title='Level')
  plt.tight_layout()
  plt.show()
```

```python
path = r"D:\1. PROJECT\Mangroves CDO\analysis - 9-30-2025\buffer\non_ahp_report.csv"
data_bargraph(path, 'Non-AHP', 'level')
```
