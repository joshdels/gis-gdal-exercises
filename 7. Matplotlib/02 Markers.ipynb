```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker='*')
plt.show()
```

marker|line|color

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o-.k')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker='o', ms=20, mec='red', mfc='r')
plt.show()
```
