```python
import numpy as np
import  pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\2. AREAS\JUPYTER\Gdal_scripts\2. projects\5 Mangroves\data\cross_over.csv')
print(df.to_string())
```

```python
mean_barangay = df.groupby("ADM4_EN")["_mean"].mean().reset_index().sort_values(by="_mean", ascending=True)
print(mean_barangay)
```

```python
plt.figure(figsize=(8,5))
plt.bar(mean_barangay['ADM4_EN'], mean_barangay["_mean"])

# Labels and title
plt.title("Mean Vulnerability of Mangroves per Barangay", fontsize=14)
plt.xlabel("Barangay", fontsize=12)
plt.ylabel("Mean Value", fontsize=12)

# Show the chart
plt.tight_layout()
plt.show()
```
