This is to calculate the clipped 100n buffer area using geopandas and also to dissolved the propeties

```python
import geopandas as gpd
```

```python
gdf = gpd.read_file("./data/non-ahp-clipped-100m.geojson").to_crs(3857)
```

```python
dissolved = gdf.dissolve(by='DN')  
```

```python
dissolved["area_m2"] = dissolved.geometry.area
```

```python
print(dissolved[["area_m2"]])
```

```python
dissolved.to_file("./data/non-ahp_dissolved.geojson", driver="GeoJSON")
print("exported complete")
```
