This is to calculate the clipped 100n buffer area using geopandas and also to dissolved the propeties

```python
import geopandas as gpd
```

```python
gdf = gpd.read_file("./data2/non-ahp_dissolved.geojson").to_crs(3857)
```

```python
dissolved = gdf.dissolve(by='class_2')  
```

```python
dissolved["area_vulnerability_m2"] = dissolved.geometry.area
```

```python
dissolved["vulnerability_has"] = dissolved["area_vulnerability_m2"]/10000
```

```python
print(dissolved[["vulnerability_has"]])
```

```python
dissolved.to_file("./data2/non_ahp_calculated_vulnerability_v2.geojson", driver="GeoJSON")
print("exported complete")
```
